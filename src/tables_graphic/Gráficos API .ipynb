```python
import pandas as pd 
```

```python
dados_sensores = pd.read_excel('dadosSensores_grafico.xlsx')
print (dados_sensores.head())
```

```python
import matplotlib.pyplot as plt 
```

```python
dia = dados_sensores['DATA']
umidade_do_ambiente = dados_sensores['UMIDADE DO AMBIENTE']
plt.figure(figsize=(7, 5))  # Definir o tamanho da figura
plt.plot(dia, umidade_do_ambiente, marker='o', color='skyblue', linestyle='-')
plt.title('Relação dia e Umidade do ambiente')
plt.xlabel('Data')
plt.ylabel('Umidade do Ambiente')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.grid(True)
plt.show()

```

```python
dia = dados_sensores['DATA']
temperatura = dados_sensores['TEMPERATURA']
plt.figure(figsize=(10, 6))  # Definir o tamanho da figura
plt.plot(dia, temperatura,marker='o', color='skyblue', linestyle='-')
plt.title('Relação dia/Temperatura')
plt.xlabel('Data')
plt.ylabel('Temperatura')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.grid(True)
plt.show()
```

```python
dia = dados_sensores['DATA']
umidade = dados_sensores['UMIDADE DO SOLO']
plt.figure(figsize=(10, 6))  # Definir o tamanho da figura
plt.plot(dia, umidade,marker='o', color='skyblue', linestyle='-')
plt.title('Relação dia/Umidade do solo')
plt.xlabel('Data')
plt.ylabel('Umidade do solo')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.grid(True)
plt.show()
```

```python
dia = dados_sensores['DATA']
volume = dados_sensores['VOLUME DE ÁGUA EM L']
plt.figure(figsize=(10, 6))  # Definir o tamanho da figura
plt.plot(dia, volume, marker='o', color='skyblue', linestyle='-')
plt.title('Relação dia e Volume de água em L')
plt.xlabel('Data')
plt.ylabel('Volume de água em L')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
plt.grid(True)
plt.show()
```
